## UVOD

Navor na tuljavo z $N$ navoji, površino $S$ in tokom $I$ v zunanjem magnetnem polju $B$ je enak $M=NIS \times B$. Magnetno polje $B$ lahko ustvarimo z drugo tuljavo in je sorazmerno s številom navojem in tokom na tuljavi. Sorazmernostni koeficient je indukcijska konstanta $ \mu_0 $, ki jo bomo merili v nalogi.

Uporabljali bomo par Helmholtzovih tuljav, tj. dve enaki okrogli zaporedno vezani tuljavi, vsaka z $N_H$ ovoji, oddaljeni pa sta za $R_H$, prav toliko znaša njun radij. Na sredini med obema je gostota magnetnega polja precej homogena:
$$ B = (4/5)^{3/2} \frac{\mu_0 N_H I_H}{R_H} $$
Navor na tuljavo je torej $M=NISB \sin{\phi}$

## NALOGA:

Izmeriti indukcijsko konstanto. Med obe Helmholtzovi tuljavi bomo pravokotno postavili manjšo tuljavo ter spreminjali tako tok skozi H. tuljavi $I_h$ kot tok skozi manjšo tuljavo $I$. S torzijskim merilnikom navorov bomo pri vsakem paru tokov uravnovesili povzročeni navor, tako da bo kot med magnetnim poljem in manjšo tuljavo vedno $\pi/2$. Iz trojic $M, I, I_H$ lahko določimo indukcijsko konstanto.
