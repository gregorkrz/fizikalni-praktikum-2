## UVOD

V električnem nihajnem krogu se pri nihanju izmenjujeta električna energija kondenzatorja in magnetna energija tuljave. Zaradi dušenja (upora žic, notranjega upora komponent) nihanje po določenem času izzveni. Če ga vzbujamo s sinusno napetostjo, nihanje ne izzveni, pač pa niha z amplitudo in faznim zamikom (izmerimo z osciloskopom), ki je odvisna od frekvence vzbujanja in je največja, ko je sistem v resonanci ($\nu_{vzbujanja} = \nu_0 $).

$ \nu_0 = \frac{1}{2 \pi \sqrt{LC}}$
