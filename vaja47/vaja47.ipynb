## UVOD

Elektrodi ploščatega kondenzatorja se privlačita zaradi električnih privlačnih sil. Kondenzator s ploščino $S$ in razmikom med ploščama $d$ ima kapaciteto $C=\epsilon_0 S / d$. Če je na ploščah napetost $U$, je sila sorazmerna z nabojem na prvi plošči $e_1$ in jakosti el. polja, ki bi ga dobili samo z drugo ploščo $E_2$: $$F=e_1E_2=\frac{\epsilon_0 S U^2}{2d^2}$$

## NALOGA
