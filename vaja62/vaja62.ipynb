## 1. Projekcijski aparat
Sestavljeni aparat je prikazan na spodnji skici. Sestavljen je iz svetila, kondenzorja in objektiva. Sliko S diapozitiva P projiciramo na zaslon. Kondenzor postavimo pred diapozitiv zato, da dobimo vzporedne žarke in je zato slika enakomerno osvetljena.
### Meritve
$$ f_k = 13 cm \pm 5 mm $$
$$ f_{obj} = (139 \pm 3) mm $$
Višina slike $ S = 7.7 cm \pm 2 mm $

Višina diapozitiva (predmeta) $ P = 3.4 cm \pm 2 mm$
$$a = 18.5 cm \pm 3 mm $$
$$b = 46.5 cm \pm 5 mm $$

Povečava

$$M_{izmerjena} = S/P \approx 2.3 \pm 0.2 $$
$$M_{izračunana} = b/a = 2.5 \pm 0.05 $$

### Meritve
Gor. razdalji preberemo z leč:
$$ f_1 = 51 cm $$
$$ f_2 = 8.7 cm $$
$$ a = 12 m \pm 2 m $$
Oddaljenost nitnega križa od ravnine okularja:
$$ q = 5.8 cm \pm 0.5 cm $$
Razdalja med dvema razdelkoma na okularju:
$$ r = (1.5 \pm 0.5) mm $$
Opazujemo predmet na razdalji a, visok $x = 2.0 cm $. Čez en razdelek se razteza pribl. $N=4$ predmetov.

Povečava
$$ M_{izračunana} = \frac{f_1}{f_2} \frac{1}{1-f_1 / a} = 6.12 \pm 0.05$$
Izmerjeno povečavo ocenimo iz razmerja zornih kotov.
$$ M_{izmerjena} = \frac{r/q}{N*x/a} \approx 3.9 \pm 1.5 $$

## 3. Mikroskop
Sestavimo mikroskop s skice spodaj. Povečavo definiramo kot razmerje tangensov zornega kota, pot katerim vidimo predmet skozi mikroskop, in zornega kota, pri katerem bi ga videli pri $r=25cm$.
$$ M = N_1 * N_2 = \frac{d r}{f_1 f_2}$$
$$N_1 = d/f_1 $$
$$N_2 = r/f_2 $$
$d $ je razdalja med notranjima goriščema obeh leč.
